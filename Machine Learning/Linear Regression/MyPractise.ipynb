```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
house = pd.read_csv('Housing.csv')
```

```python
house.head()
```

```python
print(house.shape)
house.describe()
```

```python
sns.pairplot(house)
plt.show()
```

```python
sns.boxplot(x='furnishingstatus', y= 'price',hue='airconditioning', data=house)
plt.show()
```

```python
varlist =  ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
house[varlist] = house[varlist].apply(lambda x: x.map({'yes':1, 'no':0}))
```

```python
status = pd.get_dummies(house['furnishingstatus'], drop_first=True )
```

```python
house = pd.concat([house, status], axis=1)
```

```python
house.drop(['furnishingstatus'], axis = 1, inplace = True)
```

```python
house.head()
```

```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(house, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
num_vars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking','price']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
df_train.head()
```

```python
plt.figure(figsize = (16, 10))
sns.heatmap(df_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

```python
y_train = df_train.pop('price')
X_train = df_train
```

```python
import statsmodels.api as sm
```

```python
X_train_lm = sm.add_constant(X_train['area'])
lr = sm.OLS(y_train, X_train_lm).fit()
```

```python
lr.summary()
```

```python
X_train_lm = sm.add_constant(X_train[['area', 'bathrooms', 'bedrooms']])
lr = sm.OLS(y_train, X_train_lm).fit()
```

```python
lr.summary()
```

```python
X_train_lm = sm.add_constant(X_train)
lr = sm.OLS(y_train, X_train_lm).fit()
```

```python
lr.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif['features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = 'VIF', ascending = False)
```

```python
vif
```

```python
X = X_train.drop('semi-furnished', 1,)
```

```python
X_train_lm = sm.add_constant(X)
lr_2 = sm.OLS(y_train, X_train_lm).fit()
```

```python
lr_2.summary()
```

```python
vif = pd.DataFrame()
vif['features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = 'VIF', ascending = False)
vif
```

```python
X = X.drop(['bedrooms','basement'], 1)
```

```python
X_train_lm = sm.add_constant(X)
lr_3 = sm.OLS(y_train, X_train_lm).fit()
```

```python
lr_3.summary()
```

```python
vif = pd.DataFrame()
vif['features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = 'VIF', ascending = False)
vif
```

```python
y_train_price = lr_3.predict(X_train_lm)
```

```python
sns.distplot((y_train-y_train_price))
```

```python
num_vars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking','price']

df_test[num_vars] = scaler.transform(df_test[num_vars])
```

```python
y_test = df_test.pop('price')
X_test=df_test
```

```python
X_test_sm = sm.add_constant(X_test)
```

```python
X_test_sm = X_test_sm.drop(["bedrooms", "semi-furnished", "basement"], axis = 1)
y_pred_sm = lr_3.predict(X_test_sm)
```

```python
y_test_pred = lr_3.predict(X_test_sm)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score = r2_score(y_test, y_test_pred)
print(r2_score)
```
