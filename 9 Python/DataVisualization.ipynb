```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1,10,100)
y = np.log(x)
plt.figure(1)

plt.subplot(221)
plt.title('y=log(x)')
plt.plot(x,y)

plt.subplot(222)
plt.title('y=log(x)**2')
plt.plot(x,y**2)
plt.show()
```

```python
prod_df = pd.read_csv('prod_dimen.csv')
```

```python
df3 = pd.merge(df, prod_df, how = 'inner', on='Prod_id')
df3.head()
```

```python
sns.boxplot(x='Product_Category', y = 'Sales', data=df3)
plt.yscale('log')
plt.show()
```

```python
sns.boxplot(x='Product_Category', y = 'Profit', data=df3)
plt.show()
```

```python
plt.figure(figsize = (10,8))
plt.subplot(2,2,1)
sns.boxplot(x='Product_Category', y = 'Sales', data=df3)
plt.yscale('log')

plt.subplot(2,2,2)
sns.boxplot(x='Product_Category', y = 'Profit', data=df3)
plt.yscale('log')

plt.show()
```

```python
plt.figure(num = None, figsize = (12,8), dpi=80, facecolor='w', edgecolor='k')
sns.boxplot(x='Product_Sub_Category', y = 'Profit', hue = 'Product_Category', data=df3)
plt.show()
```

```python
sns.barplot(x='Product_Category', y = 'Sales', data=df3)
plt.show()
```

```python
plt.figure(figsize = (10,8))
plt.subplot(2,2,1)
sns.barplot(x='Product_Category', y = 'Sales', data=df3)

plt.subplot(2,2,2)
sns.barplot(x='Product_Category', y = 'Profit', data=df3)

plt.show()
```

```python
sns.barplot(x='Profit', y = 'Product_Sub_Category', hue = 'Product_Category', data=df3, estimator=np.median)
plt.show()
```
